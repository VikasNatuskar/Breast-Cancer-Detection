```python
train = cancer_df.drop(['target'], axis=1)
test = cancer_df['target']
```

## Split train and test Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, train_size= 0.7, random_state=123)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
y_train
```

## Build model, Train model and Test model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

### LOGISTIC REGRESSION

```python
lr_ = LogisticRegression()
lr_.fit(x_train, y_train)
y_pred_lr = lr_.predict(x_train)
print("Training Accuracy_score = ", np.round(accuracy_score(y_true= y_train, y_pred= y_pred_lr)* 100,2))
```

```python
y_pred_lr2 = lr_.predict(x_test)
print("Testing Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_lr2)* 100,2))
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
svc_classifier = SVC()
svc_classifier.fit(x_train, y_train)
y_pred_svc = svc_classifier.predict(x_train)
print("Training Accuracy_score = ", np.round(accuracy_score(y_true= y_train, y_pred= y_pred_svc)* 100,2))
```

```python
y_pred_svc2 = svc_classifier.predict(x_test)
print("Testing Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_svc2)* 100,2))
```

### K - Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn_classifier = KNeighborsClassifier(n_neighbors= 5, metric='minkowski', p = 2)
kn_classifier.fit(x_train, y_train)
y_pred_knc = kn_classifier.predict(x_train)
print("Training Accuracy_score = ", np.round(accuracy_score(y_true= y_train, y_pred= y_pred_knc)* 100,2))
```

```python
y_pred_knc2 = kn_classifier.predict(x_test)
print("Testing Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_knc2)* 100,2))
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_classifier = GaussianNB()
nb_classifier.fit(x_train, y_train)
y_pred_nb = nb_classifier.predict(x_train)
print("Training Accuracy_score = ", np.round(accuracy_score(y_true= y_train, y_pred= y_pred_nb)* 100,2))
```

```python
y_pred_nb2 = nb_classifier.predict(x_test)
print("Testing Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_nb2)* 100,2))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_classifier = DecisionTreeClassifier(criterion= 'entropy', max_depth= 100)
tree_classifier.fit(x_train, y_train)
y_pred_tree = tree_classifier.predict(x_test)
print("Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_tree)* 100,2))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators= 100, max_depth= 10, criterion = 'entropy')
rf_classifier.fit(x_train, y_train)
y_pred_rf = rf_classifier.predict(x_test)
print("Accuracy_score = ", np.round(accuracy_score(y_true= y_test, y_pred= y_pred_rf)* 100,2))
```

### XGBoost Classifier

```python
from xgboost import XGBClassifier
```

```python
xgb_classifier = XGBClassifier(n_estimators = 100, max_depth = 10, learning_rate = 0.02)
xgb_classifier.fit(x_train, y_train)
y_pred_xgb = xgb_classifier.predict(x_test)
print("Accuracy_score = ", np.round(accuracy_score( y_true= y_test, y_pred= y_pred_xgb )* 100,2))
```
